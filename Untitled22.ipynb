```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
troubled = pd.read_csv('https://raw.githubusercontent.com/Vilars/Data-205/refs/heads/main/Troubled.csv', index_col=0)
troubled.head(5)
```

```python
troubled.describe()
```

```python
troubled.isnull().sum()
```

```python
troubled = troubled.dropna()
```

```python
# fix city name
troubled['City'] = troubled['City'].str.title()
```

```python
troubled['Severity Index'] = troubled['Severity Index'].replace(-96.25, np.nan)
```

```python
# distribution of average violation per unit
plt.figure(figsize=(10, 6))
sns.histplot(troubled['Average Violations Per Unit'], bins=20, kde=True)
plt.xlabel('Average Violations Per Unit')
plt.ylabel('Frequency')
plt.title('Distribution of Average Violations Per Unit')
plt.xlim(0, 20)
plt.tight_layout()
plt.show()
```

```python
# top 10 city
troubled['City'].value_counts().head(10)

# barplot top 10 city
sns.countplot(x='City', data=troubled, order=troubled['City'].value_counts().head(10).index)
plt.xticks(rotation=90)
plt.show()
```

```python
# Boxplot analysis rating vs average violation
sns.boxplot(x='Analysis Rating', y='Average Violations Per Unit', data=troubled)
plt.xlabel('Analysis Rating')
plt.ylabel('Average Violations per Unit')
plt.title('Boxplot of Rating vs Average Violations Per Unit')
plt.show()
```

```python
# Boxplot analysis rating vs severity index
sns.boxplot(x='Analysis Rating', y='Severity Index', data=troubled)
plt.xlabel('Analysis Rating')
plt.ylabel('Severity Index')
plt.title('Boxplot of Rating vs Severity Index')
plt.show()
```

```python
# Boxplot analysis rating vs unit count
sns.boxplot(x='Analysis Rating', y='Unit Count', data=troubled)
plt.xlabel('Analysis Rating')
plt.ylabel('Unit Count')
plt.title('Boxplot of Rating vs Unit Count')
plt.show()
```

```python
# Heatmap
corr = troubled.select_dtypes(include=np.number).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
crime = crime.dropna(subset=['Zip Code'])
```

```python
#merge
df = pd.merge(troubled, crime, left_on='Zip Code', right_on='Zip Code', how='left')
df.head()

```

```python
# Count the number of crimes near each property
crime_counts = df.groupby('Zip Code').size().reset_index(name='Crime Count')

# Merge the crime counts with the troubled properties data
troubled_properties_with_crime = pd.merge(troubled, crime, on='Zip Code', how='left').fillna(0)

# Compare the severity index of property issues with the frequency and types of crimes
severity_vs_crime = troubled_properties_with_crime[['Zip Code', 'Severity Index', 'Crime Count']]

# Display the final dataframe with severity index and crime counts
print("\nSeverity Index vs Crime Counts:")
print(severity_vs_crime.head())
```

```python
# convert zip code in string
troubled['Zip Code'] = troubled['Zip Code'].astype(str)
crime['Zip Code'] = crime['Zip Code'].astype(str)

#merge
df = pd.merge(crime, troubled, left_on='Zip Code', right_on='Zip Code', how='inner')
df.head()

# Count the number of crimes near each property
crime_counts = df.groupby('Zip Code').size().reset_index(name='Crime Count')

# Merge the crime counts with the troubled properties data
troubled_properties_with_crime = pd.merge(troubled, crime_counts, on='Zip Code', how='left').fillna(0)

# Compare the severity index of property issues with the frequency and types of crimes
# Use 'Crime Count' instead of 'Crime Counts'
severity_vs_crime = troubled_properties_with_crime[['Zip Code', 'Severity Index', 'Crime Count']]

# Display the final dataframe with severity index and crime counts
print("\nSeverity Index vs Crime Counts:")
print(severity_vs_crime.head())
```

```python
# Group and aggregate
grouped_data = df.groupby('Zip Code')[['Crime Name1', 'No Violations Observed']].agg(lambda x: x.value_counts().index[0]) # replace with your actual columns
# Correlation matrix
correlation_matrix = grouped_data.corr()
# Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Crime Types and Property Violations')
plt.show()
# Scatter plot example
plt.scatter(grouped_data['Crime Name1'], grouped_data['No Violations Observed']) # replace with your actual columns
plt.xlabel('Crime Type')
plt.ylabel('No Violations Observed')
plt.title('Scatter Plot of Crime Type vs. Property Violation Type')
plt.show()
```
